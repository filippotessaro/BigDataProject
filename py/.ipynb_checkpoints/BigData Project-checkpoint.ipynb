```python
from pyspark.sql.functions import concat, col, lit
location_content = requests.get('https://s3.amazonaws.com/nyc-tlc/misc/taxi+_zone_lookup.csv').iter_lines()
location_translate_schema = spark.read.csv(sc.parallelize(location_content), inferSchema = 'true', header = 'true', nullValue = 'null')
location_translate_schema.createOrReplaceTempView("location")
location_translate_schema.printSchema

test = spark.createDataFrame(taxi_schema.take(1000))

#we need before clean up the dataset from the null places (264 - 265)
t = test.filter(test['PULocationID'] != 264)
#t.createOrReplaceTempView("record")

fromTo = t.select(concat(col("PULocationID"), lit("-"), col("DOLocationID")))
#print len(fromTo.columns)
fromTo = fromTo.withColumnRenamed('concat(PULocationID, -, DOLocationID)', "fromTo")

#most important track
display(fromTo.groupBy('fromTo').count().orderBy('count', ascending=False))
```
